```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/users/gr_1/downloads/Figure5_poster_data.csv')
```

```python
for i in range(len(df.columns)-1):
    column = df.columns[i]
    plt.figure(figsize=(13,5))
    sns.scatterplot(data=df,x=column,y='Denudation rate (mm/kyr)')
```

```python
for i in range(len(df.columns)):
    column = df.columns[i]
    df[column + 'quantile'] = pd.qcut(df[column],7)
    
    plt.figure(figsize=(13,5))
    sns.boxplot(data=df,x=column +'quantile',y='Denudation rate (mm/kyr)')
```
